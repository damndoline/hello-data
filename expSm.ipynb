```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.plotly as py
import plotly.figure_factory as ff
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
py.plotly.tools.set_credentials_file(username='dopf', api_key='oTdrjRWtB9H2MWm9w98V')
```

```python
N = 100
x_t = np.arange(N)
y_rand = np.random.rand(N)
```

```python
alpha = 0.1 #smoothing factor

y_sm = np.zeros(100) #empty array to hold smoothed values

it = np.nditer(y_rand, order='C', flags=['f_index'])
while not it.finished:
    #print("%f <%d> \n" % (it[0], it.index), end=' ')
    if it.index == 0:
        y_sm[it.index] = it[0]
    else:
        y_sm[it.index] = alpha * it[0] + (1-alpha) * y_sm[it.index-1]
    it.iternext()
```

```python
trace1 = go.Scatter(
    x = x_t,
    y = y_rand
)

trace2 = go.Scatter(
    x = x_t,
    y = y_sm
)

data = [trace1,trace2]

layout = go.Layout(
    xaxis=dict(
        range=[0, 100]
    ),
    yaxis=dict(
        range=[-2, 2]
    )
)

fig = go.Figure(data=data, layout=layout)

py.iplot(fig, data=data, layout=layout)
```
